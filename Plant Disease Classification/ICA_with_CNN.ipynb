```python
# Plant Disease Detection using ICA + CNN

import os
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA, FastICA
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Conv2D, MaxPooling2D,
                                     BatchNormalization, GlobalAveragePooling2D,
                                     Dense, Dropout)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

from google.colab import drive
drive.mount('/content/drive')

data_dir = "/content/drive/MyDrive/Plant_Disease_Classification-main/plant_disease_mini"

img_size = 64
X, y = [], []

class_names = sorted(os.listdir(data_dir))
class_to_idx = {cls: idx for idx, cls in enumerate(class_names)}

print("Classes found:", class_names)

for cls in class_names:
    folder = os.path.join(data_dir, cls)
    if not os.path.isdir(folder):
        continue
    print(f"Loading class: {cls}")
    for fname in os.listdir(folder):
        fpath = os.path.join(folder, fname)
        try:
            img = load_img(fpath, target_size=(img_size, img_size))
            arr = img_to_array(img) / 255.0      # scale to [0,1]
            X.append(arr)
            y.append(class_to_idx[cls])
        except Exception as e:
            # skip unreadable files
            print("Skipping file:", fpath, "because:", e)
            continue

X = np.array(X, dtype=np.float32)
y = np.array(y, dtype=np.int64)

print("Dataset shape:", X.shape, "Labels:", y.shape)

X_flat = X.reshape(len(X), -1)
print("Raw feature dimension:", X_flat.shape[1])


n_comp = 50
pca_whiten = PCA(n_components=n_comp, whiten=True, random_state=42)
X_white = pca_whiten.fit_transform(X_flat)
print("Whitened dimension:", X_white.shape[1])

ica = FastICA(n_components=n_comp,
              random_state=42,
              max_iter=400,
              tol=1e-4)
X_ica = ica.fit_transform(X_white)
print("ICA component shape:", X_ica.shape)


X_white_recon = np.dot(X_ica, ica.mixing_.T)


X_recon_flat = pca_whiten.inverse_transform(X_white_recon)


X_recon = X_recon_flat.reshape(-1, img_size, img_size, 3)
X_recon = np.clip(X_recon, 0.0, 1.0)

print("Reconstructed image shape:", X_recon.shape)



X_train, X_test, y_train, y_test = train_test_split(
    X_recon, y, test_size=0.2, random_state=42, stratify=y
)

print("Train set:", X_train.shape, "Test set:", X_test.shape)


num_classes = len(class_names)

def build_cnn(input_shape=(64, 64, 3), n_classes=4):
    model = Sequential([
        Conv2D(32, (3,3), activation='relu', padding='same',
               input_shape=input_shape),
        BatchNormalization(),
        MaxPooling2D((2,2)),

        Conv2D(64, (3,3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2,2)),

        Conv2D(128, (3,3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2,2)),

        GlobalAveragePooling2D(),
        Dense(128, activation='relu'),
        Dropout(0.4),
        Dense(n_classes, activation='softmax')
    ])
    return model

model = build_cnn(input_shape=(img_size, img_size, 3),
                  n_classes=num_classes)

model.compile(optimizer=Adam(learning_rate=3e-4),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()



es = EarlyStopping(monitor='val_loss',
                   patience=6,
                   restore_best_weights=True,
                   verbose=1)

history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2,
    callbacks=[es],
    verbose=1
)



plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss (ICA + CNN)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title('Accuracy (ICA + CNN)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()


y_prob = model.predict(X_test)
y_pred = np.argmax(y_prob, axis=1)

print("Test Accuracy (ICA + CNN):", accuracy_score(y_test, y_pred))
print("\nClassification report:\n",
      classification_report(y_test, y_pred, target_names=class_names))


ica_vis = FastICA(n_components=2, random_state=42, max_iter=400, tol=1e-4)
X_ica_vis = ica_vis.fit_transform(X_white)

plt.figure(figsize=(7,5))
for idx, cls in enumerate(np.unique(y)):
    plt.scatter(X_ica_vis[y==cls,0],
                X_ica_vis[y==cls,1],
                alpha=0.6,
                label=class_names[idx])
plt.legend()
plt.title("2D ICA projection of leaf classes")
plt.xlabel("IC1")
plt.ylabel("IC2")
plt.show()
```
